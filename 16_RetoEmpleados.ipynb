```python
#2 Leer CSV
EmpleadosAttrition = pd.read_csv ('/content/gdrive/MyDrive/Colab Notebooks/Data15/empleadosRETO.csv')
EmpleadosAttrition
```

```python
#3 Eliminar columnas
EmpleadosAttrition.drop(['EmployeeCount', 'Over18', 'StandardHours'], axis=1 ,inplace=True)


```

```python
EmpleadosAttrition.dtypes
```

```python
#4 Crear columnas
#año de contratación
EmpleadosAttrition['fechaTmp'] = pd.to_datetime(EmpleadosAttrition['HiringDate'], errors='coerce')

EmpleadosAttrition.dropna(subset=['fechaTmp'], inplace=True)
EmpleadosAttrition['Year'] = EmpleadosAttrition['fechaTmp'].dt.year
EmpleadosAttrition['Year'] = EmpleadosAttrition['Year'].astype(int)
EmpleadosAttrition.drop(['fechaTmp'], axis=1 ,inplace=True)
#b Años que lleva en la compañia
EmpleadosAttrition['YearsAtCompany'] = 2018 -  EmpleadosAttrition.Year
```

```python
#Columna de distancia en numérico
EmpleadosAttrition.rename(columns={'DistanceFromHome':'DistanceFromHome_km'}, inplace=True)
EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome_km'].str.replace(' km', '', regex=False)
EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome'].astype(int)
```

```python
#6 Eliminar las columnas innecesarias
EmpleadosAttrition.drop(['Year','HiringDate','DistanceFromHome_km'], axis=1 ,inplace=True)
```

```python
EmpleadosAttrition
```

```python
#7 Ingreso promedio por departamentos
SueldoPromedioDepto = EmpleadosAttrition.groupby('Department')['MonthlyIncome'].mean().reset_index()
SueldoPromedioDepto.rename(columns={'MonthlyIncome':'SueldoPromedio'}, inplace=True)
SueldoPromedioDepto
```

```python
#8 Escalar la columna [MonthlyIncome]
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
EmpleadosAttrition['MonthlyIncome_Scarlar'] = scaler.fit_transform(EmpleadosAttrition[['MonthlyIncome']])
EmpleadosAttrition
```

```python
#9 Convertir las columnas [BusinessTravel,Department,EducationField,Gender,JobRole,MaritalStatus,Attrition]
from sklearn.preprocessing import OneHotEncoder

columnas = ['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Attrition']

for col in columnas:
    # limpiar los nulos
    EmpleadosAttrition[col] = EmpleadosAttrition[col].astype(str).fillna('')
    encoder = OneHotEncoder()
    encoded_data = encoder.fit_transform(EmpleadosAttrition[[col]])

    encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out([col]), index=EmpleadosAttrition.index)
    EmpleadosAttrition = pd.concat([EmpleadosAttrition, encoded_df], axis=1)


EmpleadosAttrition  #58 columnas
```

```python
#10 Correlación lineal de cada una de las variables

#eliminamos las variables categoricas

df = EmpleadosAttrition.drop(['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Attrition','OverTime'], axis=1 ) #,inplace=True
df_matrix = df.corr()
df_matrix
```

```python
#10 seleccionar variables que tengan una correlación mayo a 0.1
#EmpleadosAttritionFinal = df_matrix[df_matrix.abs() > 0.1]
EmpleadosAttritionFinal = df_matrix[df_matrix.abs() > 0.1].dropna(axis=0, how='all').dropna(axis=1, how='all')
EmpleadosAttritionFinal
```
